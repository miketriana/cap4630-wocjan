# Problem 1
Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that

* p1 is a minimum of f1
* p2 is a maximum of f2
* p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector.

This first step is to import the `mplot3d` toolkit from Matplotlib:


```
from mpl_toolkits import mplot3d
```

We will also import numpy and pyplot:

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# Function 1
We will define the following function as f1:

$f1(x, y) = x^2 + y^2$

```
def f1(x, y):
    return x ** 2 + y ** 2
```

We will then create a three-dimensional plot to visualize this function. The data for the plot will be a range of -6 to 6 for the x and y axes, while the z axis will be the value of f1 at each point.

```
xs = np.linspace(-6, 6, 30)
ys = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(xs, ys)
Z = f1(X, Y)
```

```
ax = plt.axes(projection="3d")
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

We need to find a point p1 such that p1 is a minimum of f1. Taking the partial derivatives of $z = f1(x, y)$:

$\frac{\partial z}{\partial x} = 2x$
$\frac{\partial z}{\partial y} = 2y$

Setting these to zero and solving for x and y, we find the critical point $(0, 0)$.

Using the second partial derivatives test, we find that

$f1_{xx}(0, 0)f1_{yy}(0, 0)-(f1_{xy}(0, 0))^2 = 4 > 0$

$f1_{xx}(0, 0) = 2 > 0$

Therefore, $(0, 0)$ is a local minimum of f1.

```
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

x = 0
y = 0
ax.scatter3D(x, y, f1(x, y));
```

Here is a visualization of the point on a density plot and contour plot:

```
plt.style.use('seaborn-white')
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.scatter(x, y);
```

```
plt.contour(X, Y, Z, 20, colors='black')
plt.scatter(x, y);
```

# Function 3
We will define the following function as f3:

$f3(x, y) = x^2 - y^2$

```
def f3(x, y):
    return x ** 2 - y ** 2
```

Plotting this function:

```
xs = np.linspace(-6, 6, 30)
ys = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(xs, ys)
Z = f3(X, Y)
```

$\frac{\partial z}{\partial x} = 2x$
$\frac{\partial z}{\partial y} = -2y$

$f3_{xx}(0, 0)f3_{yy}(0, 0)-(f3_{xy}(0, 0))^2 = -4 < 0$

$\nabla f3(0,0) = \begin{bmatrix}0 \\ 0\end{bmatrix}$
